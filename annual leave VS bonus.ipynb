```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

```python
 df = pd.read_excel('annual_bonus.xlsx')
```

```python
df.head()
```

```python
#Scatterplot to analyse the relationship between annual leave taken and bonus
plt.figure(figsize=(10, 6))
for job_title in df['JobTitle'].unique():
    plt.scatter(df[df['JobTitle'] == job_title]['VacationHours'], 
                df[df['JobTitle'] == job_title]['Bonus'], 
                label=job_title)

plt.title('Scatter Plot of Vacation Hours vs Bonus')
plt.xlabel('Vacation Hours')
plt.ylabel('Bonus')
plt.legend()
plt.grid(True)
plt.show()
```

```python
# Group by JobTitle and calculate average Bonus for each VacationHours level
grouped_df = df.groupby(['JobTitle', 'VacationHours'])['Bonus'].mean().reset_index()

# Pivot table for plotting
pivot_df = grouped_df.pivot(index='VacationHours', columns='JobTitle', values='Bonus')

# Plot grouped bar plot
pivot_df.plot(kind='bar', figsize=(12, 6))
plt.title('Average Bonus by Job Title and Vacation Hours')
plt.xlabel('Vacation Hours')
plt.ylabel('Average Bonus')
plt.xticks(rotation=45)
plt.legend(title='Job Title', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.tight_layout()
plt.show()
```
